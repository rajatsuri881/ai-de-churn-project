```python
import pandas as pd

# Load the data again (assume you're in the same directory structure)
df = pd.read_csv("data/WA_Fn-UseC_-Telco-Customer-Churn.csv")
```

```python
df.columns = df.columns.str.strip()
```

```python
# Check the column type
print(df['TotalCharges'].dtype)
```

```python
# Try converting to numeric
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
# Look at rows where conversion failed
df[df['TotalCharges'].isna()]
```

```python
#Replace NaNs with 0
df['TotalCharges'].fillna(0, inplace=True)
```

```python
# Check for duplicates
print("Duplicate rows:", df.duplicated().sum())

# Check missing values
print(df.isnull().sum())

# Confirm data types again
df.info()
```

```python
# Save to CSV
df.to_csv("data/cleaned_telco_churn.csv", index=False)
```

- Stripped whitespace from column names
- Converted 'TotalCharges' to numeric
- Handled missing values (rows with `tenure == 0`)
